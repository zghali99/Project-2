```python
# Import Dependencies
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
# Get the Netflix URL
netflix_url = "https://www.finder.com/netflix-movies"
```

```python
# Parse through the data for Netflix
html = requests.get(netflix_url).text
soup = BeautifulSoup(html, "html.parser")
```

```python
# Read the HTML page
table = soup.find_all("table")[0]
df = pd.read_html(str(table))[0]
```

```python
# Create a dataframe of movie Titles, Year of release, Runtimes and Genres for Netflix
movie_df = df["Title"]
year = df["Year of release"]
all_df = df.assign(year = year)
fin_df = all_df.drop(["year"], axis=1)
fin_df.head()
```

```python
columns = ["Title", "Year of release", "Rating", "IMDB Score", "Rotten Tomatoes Critics Score"]
empty_df = pd.DataFrame(columns=columns)

for x in range(0,1550,50):
    hulu_url = "https://reelgood.com/movies/source/hulu?filter-sort=4&offset=" + str(x)
    hulu_req = requests.get(hulu_url).text
    soup_2 = BeautifulSoup(hulu_req, "html.parser")
    table_2 = soup_2.find_all("table")[0]
    df_2 = pd.read_html(str(table_2))[0]
    hulu_movie = df_2["Title"]
    hulu_df = df_2.assign(movie = hulu_movie)
    
    hulu_df_2 = hulu_df.drop(["Unnamed: 0", "Unnamed: 2", "Unnamed: 8", "Available On", "Unnamed: 9", "movie"], axis=1)
    hulu_df_3 = hulu_df_2.rename(columns={"Year": "Year of release", "Unnamed: 5": "IMDB Score", "Unnamed: 6": "Rotten Tomatoes Critics Score"})

    hulu_df_4 = empty_df.append(hulu_df_3, ignore_index=True)
    empty_df = hulu_df_4
empty_df.head()
```

```python
fin_df["Netflix"] = "Netflix"
fin_df.head()
```

```python
# Data Cleansing -- This is the Final Neflix Dataframe before the merge
netflix_df_final = fin_df.drop(["Runtime (mins)", "Genres"], axis=1)
netflix_df_final.head()
```

```python
empty_df["Hulu"] = "Hulu"
empty_df.head()
```

```python
# Data Cleansing -- This is the Final Hulu Dataframe before the merge
hulu_df_final = empty_df.drop(["Rating", "IMDB Score", "Rotten Tomatoes Critics Score"], axis=1)
hulu_df_final.head()
```

```python
# Append both Data Frames
new_df = pd.merge(hulu_df_final, netflix_df_final, on="Title", how="left")
new_df.head()
```

```python
# Capture the movie(s) in common by dropping suplicates
result_df = pd.concat([netflix_df_final, hulu_df_final], ignore_index=True, sort=False).drop_duplicates()
result_df
```
